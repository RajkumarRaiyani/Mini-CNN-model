```python
import tensorflow as tf
import onnxmltools
```

```python
#Exporting lenetmodel into ONNX format(method-1)

IMG_SIZE_H = 28 
IMG_SIZE_W = 28
NUM_CHANNELS = 1
import tf2onnx
import onnx
input_signature = (tf.TensorSpec((None , IMG_SIZE_H, IMG_SIZE_W , NUM_CHANNELS ), tf.float32, name="image_input"),)
onnx_model, _ = tf2onnx.convert.from_keras(Newlenet5model,input_signature)
onnxmltools.utils.save_model(onnx_model, "lenetmodel.onnx")
```
